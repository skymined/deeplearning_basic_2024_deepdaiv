## 자료형

### 숫자형

숫자형 데이터로 우리가 생각하는 웬만한 계산들을 해낼 수 있습니다. 일반적인 계산기처럼 덧셈(+), 뺄셈(-), 곱셈(*), 나눗셈(/)을 모두 사용할 수 있습니다. 함수를 활용하면 더 다양한 연산을 할 수 있습니다.

특수하지만 생각보다 자주 쓰는 연산자들도 있습니다.

- 제곱 (**)
- 나눗셈 몫(//)
- 나눗셈 나머지(%)

### 문자형

처음 공부하시는 분들은 변수명과 문자형 데이터를 헷갈리기 쉽습니다. 하지만 문자형 데이터에는 따옴표가 들어가야 합니다. 함수 안에 매개변수로 값을 입력할 때는 주로 따옴표를 넣어야 합니다.

문자형 데이터도 연산이 가능합니다.

`len()` 함수를 활용하여 문자열의 길이를 셀 수 있습니다.

### 리스트


리스트에 값들을 모아서 관리할 수 있습니다. 리스트를 만들 때는 대괄호(`[]`)를 활용합니다.

리스트의 길이를 확인해봅시다. 문자열과 동일하게 `len()` 함수를 활용할 수 있습니다.

리스트도 덧셈과 곱센 연산이 가능합니다.

### 튜플
튜플도 값들의 모음입니다. 리스트와 차이점이 있다면 변경이 불가능하다는 것입니다. 변경되면 안 되는 중요한 정보를 묶을 때 활용합니다. 튜플을 만들 때는 소괄호(`()`)를 활용합니다.

# 라이브러리 활용

라이브러리는 특정 기능들을 모아놓은 코드의 집합입니다. 이미 제대로 작동하도록 구성된 여러 함수 / 모듈을 모아둔 만큼 편리하게 기능을 구현하는 데 많은 도움을 줍니다. 다양한 라이브러리가 있지만 그중에서도 정적 크롤링을 위한 `requests`와 `BeautifulSoup`를 활용해보도록 하겠습니다.

`from` 키워드는 특정 모듈이나 패키지 내부의 한 부분만을 가져올 때 사용됩니다. `from` 뒤에는 모듈이나 패키지의 이름이 오고, `import` 다음에는 그 모듈이나 패키지에서 가져오고 싶은 특정 함수, 클래스, 변수 등의 이름이 옵니다.

```python
# BeautifulSoup 클래스와 requests 모듈을 임포트
from bs4 import BeautifulSoup
import requests

url = "https://www.yes24.com//Mall/Main/Book/001?CategoryNumber=001"
```

```python
# if문 활용하기
response = requests.get(url)
if response.status_code == 200 :
    print('아싸')
else : 
    print("힝")
```

```python
# 응답 내용을 BeautifulSoup 객체로 파싱합니다.
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# HTML 문서에서 첫 번째 title 태그를 찾고 그 내용을 추출합니다.
soup
```

```python
# 크롤링을 하다가 다음과 같은 상황을 마주했다고 가정해봅시다.

a_list = soup.find_all('a','bgCateM')
part_url_list = list(map(lambda x: x.get('href'), a_list))
print(part_url_list)
```

```python
# 반복문 / 조건문 / 덧셈 활용해서 원하는 url을 추출해봅시다.
url_list =[]
full_url = 'http://www.yes24.com/'
for part_url in part_url_list :
    if '/Product/Goods' in part_url :
        url = full_url + part_url
        url_list.append(url)
        
```

# [복습 과제] 파리 올림픽 뉴스 다루기

1. https://m.sports.naver.com/paris2024/news 링크를 `url` 변수에 입력합니다.
2. if문을 활용해, `status_code`가 200일 때 '성공', 아닌 경우 '실패'를 출력합니다.
3. title_list / url_list 의 길이를 확인하고 길이가 같은 경우(`if`문 활용) `zip()` 함수로 묶어 딕셔너리(Dictionary) 타입으로 저장하고, 출력합니다.

```python
# 1번 코드
url = "https://search.naver.com/search.naver?ssc=tab.news.all&where=news&sm=tab_jum&query=%ED%8C%8C%EB%A6%AC+%EC%98%AC%EB%A6%BC%ED%94%BD"

# 2번 코드
import requests
response = requests.get(url)
if response.status_code == 200:
    print("성공><")
else :
    print("실패하셨습니다.")
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
a_list = soup.find_all('a','news_tit')

title_list = list(map(lambda x: x.get('title'), a_list))
url_list = list(map(lambda x: x.get('href'), a_list))

print(title_list)
print(url_list)
```

```python
# 3번 코드
len_title = len(title_list)
len_url = len(url_list)

if len_title == len_url :
    title_and_url = dict(zip(title_list, url_list))
    print(title_and_url)
else:
    print('무언가 잘못되었다.')
```
