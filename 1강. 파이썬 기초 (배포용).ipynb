### 숫자형

숫자형 데이터로 우리가 생각하는 웬만한 계산들을 해낼 수 있습니다. 일반적인 계산기처럼 덧셈(+), 뺄셈(-), 곱셈(*), 나눗셈(/)을 모두 사용할 수 있습니다. 함수를 활용하면 더 다양한 연산을 할 수 있습니다.

### 딕셔너리
딕셔너리 자료형은 키(Key)와 값(Value)의 쌍으로 이루어져 있습니다. 그리고 쿼리(Query)로 인덱싱이 가능하다는 특징이 있습니다. 딕셔너리는 중괄호(`{}`)를 활용합니다.

```python
member_dict = {'이딥다': 26,'김이브': 25,'박점':23}
print(member_dict['이딥다'])
```

```python
# dict() 내장함수를 이용해서 간단하게 만들 수도 있습니다.
member1 = ('이딥다', 26)
member2 = ('김이브', 25)
member3 = ('박점', 23)

age_dict = dict([member1, member2, member3]); age_dict
```

```python
# list의 길이가 동일할 때, zip() 함수를 활용해서 다음과 같이 만들 수도 있습니다.

member = ['이딥다', '김이브', '박점']
age = [26, 25, 23]

age_dict = dict(zip(member, age)); age_dict
```

## 제어문

### if
if는 말 그대로 조건문입니다. 만약 조건식이 참이라면 아래의 내용을 실행하라는 제어문입니다.

```python
a = 3
b = 2

if a + b > 4 : # if (조건식) : 조건식이 참이므로 아래의 코드를 실행
	print(a + b) # 제어문이 실행할 코드는 한 칸 들여쓰기 합니다.
```

### while
while은 반복문입니다. 조건식이 참일 경우 아래의 내용을 계속 실행합니다.

```python
a = 3
b = 2
c = a + b

while c > 1 : # while (조건식) : 조건식이 참(True)이므로 아래의 코드를 반복 실행
	print(c)
	c = c - 1 # 간단하게 c -= 1로 표시하기도 합니다.
```

### for
for도 반복문입니다. 하지만 while과 다르게 리스트 내의 원소 하나씩 실행합니다. 리스트의 원소만큼 실행됩니다.

```python
member_list = ['이딥다', '김이브', '박점']

for member in member_list: # for (원소) in (리스트):
	print(member)
```

# 라이브러리 활용

라이브러리는 특정 기능들을 모아놓은 코드의 집합입니다. 이미 제대로 작동하도록 구성된 여러 함수 / 모듈을 모아둔 만큼 편리하게 기능을 구현하는 데 많은 도움을 줍니다. 다양한 라이브러리가 있지만 그중에서도 정적 크롤링을 위한 `requests`와 `BeautifulSoup`를 활용해보도록 하겠습니다.

`from` 키워드는 특정 모듈이나 패키지 내부의 한 부분만을 가져올 때 사용됩니다. `from` 뒤에는 모듈이나 패키지의 이름이 오고, `import` 다음에는 그 모듈이나 패키지에서 가져오고 싶은 특정 함수, 클래스, 변수 등의 이름이 옵니다.

```python
# BeautifulSoup 클래스와 requests 모듈을 임포트합니다.
from bs4 import BeautifulSoup  # bs4 패키지에서 BeautifulSoup 클래스를 임포트합니다. 이 클래스는 HTML과 XML 문서를 파싱하고 탐색하는 데 사용됩니다.
import requests  # requests 모듈을 임포트합니다. 이 모듈은 HTTP 요청을 보내는 데 사용되며, 웹 콘텐츠에 접근할 때 필수적입니다.

# 웹 페이지의 URL을 지정합니다.
url = "https://www.yes24.com//Mall/Main/Book/001?CategoryNumber=001" # 스크래핑할 웹 페이지의 URL을 문자열로 저장합니다.
```

```python
# if문 활용하기

response = requests.get(url) # requests.get 메소드를 사용하여 해당 URL에 HTTP GET 요청을 보냅니다. 서버로부터의 응답은 response 객체에 저장됩니다.
if response.status_code == 200:
    print("웹 페이지를 성공적으로 가져왔습니다.")
else:
    print("웹 페이지를 가져오는 데 실패했습니다.")
```

```python
# 응답 내용을 BeautifulSoup 객체로 파싱합니다.
soup = BeautifulSoup(response.text, 'html.parser')  # 응답 객체에서 바이트 형태의 HTML 콘텐츠를 가져와 BeautifulSoup 객체를 생성합니다. 'html.parser'는 내장 HTML 파서를 사용하겠다는 의미입니다.

# 이렇게 객체를 생성하고 나면, 여러 메소드(method)와 속성(attribution)을 활용할 수 있습니다.
```

```python
# HTML 문서에서 첫 번째 title 태그를 찾고 그 내용을 추출합니다.
title = soup.find('title').text
print(title)
```

```python
# 크롤링을 하다가 다음과 같은 상황을 마주했다고 가정해봅시다.

a_list = soup.find_all('a','bgCateM')
part_url_list = list(map(lambda x: x.get('href'), a_list))
print(part_url_list)
```

```python
# 반복문 / 조건문 / 덧셈 활용해서 원하는 url을 추출해봅시다.
url_list = []
full_url = 'https://www.yes24.com'
for part_url in part_url_list:
    if '/Product/Goods' in part_url:
        url = full_url + part_url
        url_list.append(url)
        print(url)
```

# [복습 과제] 파리 올림픽 뉴스 다루기

1. https://search.naver.com/search.naver?ssc=tab.news.all&where=news&sm=tab_jum&query=파리올림픽 링크를 `url` 변수에 입력합니다.
2. if문을 활용해, `status_code`가 200일 때 '성공', 아닌 경우 '실패'를 출력합니다.
3. title_list / url_list 의 길이를 확인하고 길이가 같은 경우(`if`문 활용) `zip()` 함수로 묶어 딕셔너리(Dictionary) 타입으로 `news_dict` 변수에 저장하고, 출력합니다.

```python
# 1번 코드
url = 'https://search.naver.com/search.naver?ssc=tab.news.all&where=news&sm=tab_jum&query=파리올림픽'

# 2번 코드
response = requests.get(url)
if response.status_code == 200:
    print("성공")
else:
    print("실패")
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
a_list = soup.find_all('a','news_tit')

title_list = list(map(lambda x: x.get('title'), a_list))
url_list = list(map(lambda x: x.get('href'), a_list))

print(title_list)
print(url_list)
```

```python
# 3번 코드
if len(title_list) == len(url_list):
    news_dict = dict(zip(title_list, url_list))
    print(news_dict)
```
